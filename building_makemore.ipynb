```python
b = {} #To calculate the frequency of bigrams
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted(b.items(), key = lambda kv: -kv[1])
```

```python
#Let's transform the dictionary above to a 2d array using pytorch
import torch
```

```python
#The 2d array has to have rows as the first character and column as the second character
#Each entry in the array corresponds to the frequency of bigrams
N = torch.zeros((28, 28), dtype=torch.int32)
```

```python
#Let us find the characters in the words
comb = sorted(list(set(''.join(words))))
comb

#In order to create the lookup table(mapping of each alphabet to a number from 0-25)
stoi = {s:i for i,s in enumerate(comb)}
stoi['<S>'] = 26 #Add <s> and <e> to the table
stoi['<E>'] = 27
stoi
```

```python
#Now using the lookup table let us insert characters into N 2d array
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1] #Index of character 1 to ix1
        ix2 = stoi[ch2] #Index of character 2 to ix2
        N[ix1, ix2] += 1

N
```

```python
#Let's try to visualise the plot
import matplotlib.pyplot as plt
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
```

```python
#In a more comprehendable way?
itos = {i:s for s,i in stoi.items()}

plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(28):
    for j in range(28):
        chsr = itos[i] + itos[j]
        plt.text(j, i, chsr, ha="center", va="bottom", color="black")
        plt.text(j, i, N[i,j].item(), ha="center", va="top", color="black")
```
